```python
from  scipy.ndimage import gaussian_filter as G
from scipy.signal import argrelextrema
import torch 
import numpy as np 

# predicted joints가 input으로 들어감.
def calc_db(keypoints, name=''):
    keypoints = np.array(keypoints).reshape(-1, 24, 3)

     # velocity의 norm (크기) 의 mean을 구한 것 같음 
    kinetic_vel = np.mean(np.sqrt(np.sum((keypoints[1:] - keypoints[:-1]) ** 2, axis=2)), axis=1)
    print(kinetic_vel)
    # Gaussian function with mean kinetic_vel and standard deviation sigma 
    kinetic_vel = G(kinetic_vel, 5)
    print(kinetic_vel)
     # argrelextrema : identifies local maxima and minima in a 1D array of size N-1 
    motion_beats = argrelextrema(kinetic_vel, np.less)
    return motion_beats, len(kinetic_vel)\
    

input = torch.randn(150, 24, 3)

print(calc_db(input)[0][0])
```
